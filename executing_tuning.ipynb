```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.legend_handler import HandlerLine2D

x = np.arange(0.001, 1.0, 0.001)
#h =1-x
h = -np.log(x)
#h = -(np.log(1-x))

line1, = plt.plot(x, h, label='h(x)', linestyle='-')

plt.legend(handler_map={line1: HandlerLine2D(numpoints=4)})

plt.xlabel('f1 score')
plt.ylabel('loss')
plt.title('loss example.')
plt.grid(False)
plt.show()
plt.savefig("plot loss.png")
```

b > 1 più peso per la recall, i falsi negativi pesano di più a denominatore

b < 1 meno peso per la recall, i falsi negtivi sono meno importanti e pesano meno a denominatore

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.legend_handler import HandlerLine2D

b = np.arange(0.1, 4, 0.1)

tp=100
fp=10
fn=20

#h =1-x
h = ((1+b**2)*tp)/((1+b**2)*tp+fp+b**2*fn)
#h = -(np.log(1-x))

line1, = plt.plot(b, h, label='h(x)', linestyle='-')

plt.legend(handler_map={line1: HandlerLine2D(numpoints=4)})

plt.xlabel('beta')
plt.ylabel('score')
plt.title('fb score example.')
plt.grid(False)
plt.show()
```
